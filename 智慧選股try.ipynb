```python
data.set_universe(market='TSE_OTC') # 水泥工業
pb = data.get('price_earning_ratio:股價淨值比')
close = data.get('price:收盤價')
close
```

```python
rsi = data.indicator('RSI', timeperiod=14)
rsi
```

```python
k, d = data.indicator('STOCH')
k
```

```python
from finlab import data
from finlab.backtest import sim
from finlab import backtest

sma28 = close.average(28)
sma90 = close.average(90)
bigk = (k>d).iloc[-1]
bigd = (d>k).iloc[-1]

entries = close > sma28
exits = close < sma90

position = entries.hold_until(exits, nstocks_limit=5, rank=-pb, takeprofit=0.3, stoploss=0.1)
backtest.sim(position, resample='D')
backtest_report = sim(position)
```
